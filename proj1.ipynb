```python
import pandas as pd
import numpy as np
pd.options.mode.chained_assignment = None

weather = pd.read_csv("weather2.csv")
flights = pd.read_csv("flights.csv")
```

```python
def readWeather(filename):
    df = pd.DataFrame.from_csv(filename)
    return df.reset_index()

df = readWeather("weather2.csv")
df2 = df

df2.iloc[:57]
```

```python
dfTotal = pd.DataFrame()
for j in range(12):
    
    month = str(j+1)
    if (j+1) < 10:
            month = '0' + month
    
    df2013 = df.iloc[(19*j):(19*(j+1))]

    df2013 = df2013.drop(['year','month','X'],axis=1)
    df2013.set_index('measure', inplace = True)
    df2013 = df2013.transpose()

    df2013 = df2013.drop(['Max.TemperatureF','Min.TemperatureF','Max.Dew.PointF','MeanDew.PointF','Min.DewpointF','Max.Humidity','Mean.Humidity',
                      'Min.Humidity','Max.Sea.Level.PressureIn','Mean.Sea.Level.PressureIn','Min.Sea.Level.PressureIn','Max.VisibilityMiles','Min.VisibilityMiles',
                      'Max.Wind.SpeedMPH'],axis=1)
    df2013.loc[:,'Events'].fillna('Clear',inplace = True)
    
    df2013 = df2013.replace('T',0)
    df2013.reset_index(inplace = True)
    df2013 = df2013.rename(columns = {'index':'day'})
    
    mylist = []
    for i in range(31):
        day = str(i+1)
        if (i+1) < 10:
            day = '0' + day
        mylist.append('2013' + '-' + month + '-' + day)
    
    df2013['year'] = mylist
    df2013 = df2013.set_index('year')
    df2013 = df2013.drop('day',axis=1)
    
    dfTotal = pd.concat([dfTotal,df2013])
#df2013 = df2013.set_index('day')

dfTotal = dfTotal.dropna()
dfTotal.insert(0,'season',np.nan,allow_duplicates=True)
#dfTotal['season'] = np.nan
dfSpring = dfTotal.loc['2013-03-20':'2013-06-20']
dfSpring['season'].fillna('Spring',inplace=True)
dfSummer = dfTotal.loc['2013-06-21':'2013-09-21']
dfSummer['season'].fillna('Summer',inplace=True)
dfFall = dfTotal.loc['2013-09-22':'2013-12-20']
dfFall['season'].fillna('Fall',inplace=True)
dfWinter = pd.concat([dfTotal.loc['2013-12-21':'2013-12-31'],dfTotal.loc['2013-01-01':'2013-03-19']])
dfWinter['season'].fillna('Winter',inplace=True)

dfTotal = pd.concat([dfSpring,dfSummer,dfFall,dfWinter])

dfTotal.to_csv('TidyWeather3.csv', sep='\t', encoding='utf-8')

dfTotal
```

```python
dfSnow = dfTotal[dfTotal['Events']!='Clear']
dfSnow = dfSnow.dropna()
#print(dfSnow)
cold_temp = pd.to_numeric(dfSnow['Mean.TemperatureF']).mean()

dfClear = dfTotal[dfTotal['Events']=='Clear']
dfClear = dfClear.dropna()
#print(dfClear)
warm_temp = pd.to_numeric(dfClear['Mean.TemperatureF']).mean()

print(cold_temp)
print(warm_temp)
```

```python
def readFlights(filename):
    df = pd.DataFrame.from_csv(filename)
    return df
    
    
# raise NotImplementedError()
flights = readFlights("flights.csv")
flights = flights.reset_index()
flights = flights.sort_values('carrier')

AAonly = flights['carrier'] == 'AA'
flightsAA = flights[AAonly]
flightsAA = flightsAA.sort_values('time_hour')

JFKonly = flightsAA['origin'] == 'JFK'
flightsAAJFK = flightsAA[JFKonly]
flightsAAJFK = flightsAAJFK.sort_values('time_hour')

BOSonly = flightsAAJFK['dest'] == 'BOS'
flightsAAJFKBOS = flightsAAJFK[BOSonly]
flightsAAJFKBOS = flightsAAJFKBOS.sort_values('time_hour')

flights = flightsAAJFKBOS
```

```python
newlist = []
for i in range(len(flights)):
    month = flights.iloc[i]['month']
    if month < 10:
        month = '0' + str(month)
    else:
        month = str(month)
    
    day = flights.iloc[i]['day']
    if day < 10:
        day = '0' + str(day)
    else:
        day = str(day)
        
    newlist.append('2013-'+month+'-'+day)

#flights.drop(['month','day','dep'])
flights['date'] = newlist
flights = flights.drop(['year','month','day'],axis=1)
flights = flights.set_index('date')


flightsSnow = flights.loc[snowMask]


flightsClear = flights.loc[clearMask]
#flightsClear
flightsSnow
```
